```python
# This workflow intends to create a normalised version of Early New High German (ENHG) charters
# of the Imperial chancery under Emperor Frederick III. (1440-1493).
# The workflow intends to offer verious steps for normalising ENHG charters with
# varying degrees of depth as used by the 
# Regesta Imperii-Regesta of Emperor Frederik III (Reg.F.III).
# Therefore, only rund those lines required for the respective purpose.
```

```python
# Load text intended for normalisation

with open ('../data/test-txt.txt', 'r', encoding='utf8') as f:
    text=f.read()
```

```python
# Transform all letters to lowercase

text = text.lower()
```

```python
# Start normalisation by targeting only certain characters.
# These characters do not usually change the meaning of the text and
# can therefore be replaced or deleted without altering the content itself.
# This line does not, however, normalise abbreviations and diacritics.

for c in text:
     if c in '/⁄·‧+': # removes all punctuation marks used in Reg.F.III charters
        text = text.replace(c, '') # note that intentionally hyphens are not replaced!
     elif c in 'ſ': # replaces all instances of <tall s> with <s>
        text = text.replace(c, 's')
     elif c in 'ꝛ': # replaces all instances of <r rotunda> with <r>
        text = text.replace(c, 'r')
     elif c in 'ʒ': # replaces all instances of <ezh> with <z>
         text = text.replace(c, 'z')
```

```python
# If you wish to normalise diacritics above vowels run this line.
# The Reg.F.III ATR model produces different diacritics (superscriped o, <´>, tremmata) above <a, o, u>.
# The accent <´> may also appear above <e>.
# By default, the ATR model does not preserve diacritics for vocalised <v, w, y>.

for c in text:
     if c in 'ůüú':
         text = text.replace(c, 'u')
     elif c in 'óöò':
         text = text.replace(c, 'o')
     elif c in 'áä':
         text = text.replace(c, 'a')
     elif c in 'é':
         text = text.replace(c, 'e')
```

```python
# This line creates a list of unique tokens
# in order to build a dictionary for normalising
# abbreviations and diacritics (if those haven't been removed yet in the previous step).
# Note that the Reg.F.III ATR only produces to kinds 
# of abbreviation markers (nasal stroke and r-shortening) that can represent
# various substrings.

tokenised_text = text.split() # creates a list of all tokens
unique_tokens = set(tokenised_text) # creates a set of unique tokens based on that list
unique_tokens_list = list(sorted(unique_tokens)) # produces an alphabetic list of unique tokens
len(unique_tokens_list) # gives a count of unique tokens
```

```python
# Now the entire list of unique tokens is written in a .txt file
# and saved in the data folder.
# From there, a dictionary can be created.

with open("../data/unique_tokens.txt", "w", encoding="utf8") as f:
    f.write(f"List of unique tokens\n\nLength: {len(unique_tokens_list)} entries\n\n{str(unique_tokens_list)}")
```
